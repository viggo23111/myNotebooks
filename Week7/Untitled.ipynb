```python
import bs4
import requests


r = requests.get('https://en.wikipedia.org/wiki/Alan_Turing')
r.raise_for_status()

r.text[:500]
```

```python
soup = bs4.BeautifulSoup(r.text, 'html.parser')
title = soup.select('title')
title
```

```python
p_elems = soup.select('p')

for el in p_elems:
    if (len(el.getText().strip()) >0):
        print(el.getText())
        break
```

```python
toc = soup.select('.toc > ul > li')
toc

for li in toc:
    print(li.getText())
```

Create a dictionary from the TOC links like: {'first link':'#this_is_the_first_link}

```python
toc_dict ={}
toc_inner = soup.select('.toc > ul > li > ul > li') 

for el in toc_inner:
    text = el.getText()
    a = el.select('a')
    val = a[0].get('href')
    key = text.split(" ",1)[1]
    key = key.split("\n")[0].lower()
    toc_dict[key]=val
toc_dict
```

#### Create a function that can take a word and look for it in the dictionary keys and then return the content from the first link that matches. Return the next p elements until next headline (h3 element).

```python
def look_up(word):
    if word in toc_dict.keys():
        link = toc_dict[word]
        h3title = soup.select(link)
       # p_elems = soup.select(link > 'p')
        content = soup.select('mw-content-text')
        body = soup.select(link)
        test = soup.find('h2',{'id':link})
        print(content)
        print(test)
        return(link)
    else:
        return('Not in list')
```
