```python
df = pd.read_csv('../../../data/titanic_train.csv')
df.set_index('PassengerId')
```

```python
df.sort_values(by=['Fare'])
```

```python
def reformatName(name):
    first_name = name.split(', ')[1].split(' ')[1].strip()
    last_name = name.split(', ')[0]
    return first_name +" "+ last_name
```

```python
reformatName("Tornquist, Mr. William Henry")
```

```python
df['Name'] = df['Name'].apply(reformatName)
df
```

```python
df['Name'] = df['Name'].str.split(' ').str[1]
df
```

```python
tickets=['350406','248706','382652','244373','345763','2649','239865']
print(tickets)
df[df['Ticket'].isin(tickets)]
```

```python
df = df[df['Age'].notna()]
df
```

```python
df = df.astype({'Age':'int32'})
df
```

```python
df[df['Age']<18]
```

```python
df = df[df['Cabin'].notna()]
df
```

C = Cherbourg; Q = Queenstown; S = Southampton

```python
df = df.replace('C','Cherbourg')
df = df.replace('Q','Queenstown')
df = df.replace('S','Southampton')
df
```
