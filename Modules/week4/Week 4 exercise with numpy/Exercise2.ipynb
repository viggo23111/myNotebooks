```python
import pandas as pd 
import numpy as np
url = 'FOLK1A.csv'
df = pd.read_csv(url,sep=';')
df['TID'] = df['TID'].map(lambda x:x[:-2]) #cut the last 2 characters
df.to_csv('demografic_cleaned.csv',header=False, index=False)
```

```python
cleaned_file = 'demografic_cleaned.csv'
```

```python
loaded_data = np.loadtxt(cleaned_file, delimiter=',', dtype=np.uint)
print(loaded_data)
```

```python
import csv
city_list = {}
city_file = '../../../../data/DKstat_bykoder.csv'
csv_data = csv.reader(open(city_file, encoding = 'utf-8'), delimiter = ';')
csv_data_list = list(csv_data)

for csv_data in csv_data_list:
    city_list[int(csv_data[0])] = csv_data[1]
    
city_list
```

```python
def summed_by_city(city_code):
    print(city_code)
    
    ld = loaded_data
    control = ld[np.isin(ld[:,0],[city_code])]
    
    # male population 
    male = control[(control[:,3] == 2020) & (control[:,1] == 1)]
    male_population = male[:,4].sum()
    
    # female population
    female = control[(control[:,3] == 2020) & (control[:,1] == 2)]
    female_popultaion = female[:,4].sum()
    
    # population
    population =  control[(control[:,3] == 2020)]
    total_population = population[:,4].sum()
    
    # pct
    male_pct = (male_population/total_population) * 100
    female_pct = (female_popultaion/total_population) * 100
    
    
    print("Male population: {}, {:0.2f}%, female population: {}, {:0.2f}%, in area {} ".format(male_population, male_pct, female_popultaion, female_pct, city_code))
    
    if total_population == 0:
        return 0
    else:
        return male_pct
```

```python
summed_by_city(751)
```

```python
def get_male_pct_and_city(city_list):
    
    male_pct_and_city = {}
    
    for city_code, city_name in city_list.items():
        male_pct_and_city[city_name] = summed_by_city(city_code)
        
    return male_pct_and_city
```

```python
male_pct_dict = get_male_pct_and_city(city_list)
```

```python
sorted_dict = dict(sorted(male_pct_dict.items(), key=lambda item: item[1], reverse=True))
print(sorted_dict)
```

```python
city_with_most_men_2020 = next(iter( sorted_dict.items()))
print(city_with_most_men_2020)
```

```python
def sum_population(city_code, year):
    ld = loaded_data
    control = ld[np.isin(ld[:,0],[city_code])]
    population =  control[(control[:,3] == year)]
    total_population = population[:,4].sum()    
    print("Sum of population in year:"+str(year))
    return total_population
```

```python
sum_population(751,2020)
```

```python
def get_population_dict(city_list, year):
    population_list = {}
    for city_code, city_name in city_list.items():
        population_list[city_code] = sum_population(city_code, year)
    return population_list
```

```python
population_list_2008 = get_population_dict(city_list, 2008)
```
