```python
# read data from csv file into 2d numpy array
import numpy as np
filename = '../../../../data/befkbhalderstatkode.csv'

bef_stats_df = np.genfromtxt(filename, delimiter=',', dtype=np.uint, skip_header=1)
dd = bef_stats_df
print(dd)
```

```python
pop_area_dict = {}
for area_code, area in neighb.items():
    mask = (dd[:,0] == 2015) & (dd[:,1] == area_code)
    population = np.sum(dd[mask][:,4])
    pop_area_dict[area] = population
print (pop_area_dict)
```

```python
sorted_dict = dict(sorted(pop_area_dict.items(), key=lambda item: item[1]))
sorted_dict
```

```python
import matplotlib.pyplot as plt
plt.bar(sorted_dict.keys(), sorted_dict.values())
plt.title('Population for area', fontsize=14)
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
plt.show()
```

```python
nordic_countries = ["Norge","Sverige","Finland","Island","Grønland"] 
pop_above_65 = 0 
year = dd[:,0]
age = dd[:,2]
country_code=dd[:,3]
for code, country in country_codes.items(): 
    if country in nordic_countries: 
        mask = (year == 2015) & (age > 65) & (country_code == code) 
        country_sum = (dd[mask][0:,4].sum()) 
        pop_above_65 += country_sum
print("nordic population above 65 years old: ",pop_above_65)
```

```python
oesterbro = dd[np.isin(dd[:,1],[2])]
oesterbro = oesterbro[(oesterbro[:,0] >= 1992) & (oesterbro[:,0] <= 2015)]

vesterbro = dd[np.isin(dd[:,1],[4])]
vesterbro = vesterbro[(vesterbro[:,0] >= 1992) & (vesterbro[:,0] <= 2015)]
```

```python
oesterbro_population_dict = {}
for year in range(1992, 2016):
    oesterbro_population = oesterbro[(oesterbro[:,0] == year)]
    oesterbro_population = oesterbro_population[0:,4].sum()
    oesterbro_population_dict[year] = oesterbro_population
oesterbro_population_dict
```

```python
vesterbro_population_dict = {}
for year in range(1992, 2016):
    vesterbro_population = vesterbro[(vesterbro[:,0] == year)]
    vesterbro_population = vesterbro_population[0:,4].sum()
    vesterbro_population_dict[year] = vesterbro_population
    
vesterbro_population_dict
```
