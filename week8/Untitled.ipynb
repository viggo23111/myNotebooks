````python
import pandas as pd
df = pd.read_csv('../../data/ds_salaries.csv')
df
````

````python
data = df[['work_year', 'experience_level', 'salary_in_usd', 'job_title','company_location','company_size']]
data
````

````python
data = pd.get_dummies(data, columns=['experience_level'])
data
````

````python
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'company locations'.
data['company_location']= label_encoder.fit_transform(data['company_location'])
data
````
