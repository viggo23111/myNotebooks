### Feature engineering
- Reduce the dataset by:
  - Make == Honda
  - Fueltype == Gasoline
  - Remove outliers. Hint: `df[column name'] < df['column name'].quantile(0.90)`
  - Only use 2 features: 'Fuel Information.Highway mpg' and 'Engine Information.Engine Statistics.Horsepower'
  - Normalize data (reduce values to be between 0 and 1)

```python
df = df[(df['Identification.Make'] =='Honda') & (df['Fuel Information.Fuel Type'] =='Gasoline')]
df.head()
```

```python
df = df[(df['Fuel Information.Highway mpg'] < df['Fuel Information.Highway mpg'].quantile(0.90))]
df.size
```

```python
df = df[['Fuel Information.Highway mpg','Engine Information.Engine Statistics.Horsepower']]
df.head()
```

```python
from sklearn import preprocessing
import pandas as pd

min_max_scaler = preprocessing.MinMaxScaler()
scaled_df = min_max_scaler.fit_transform(df.values)
df = pd.DataFrame(scaled_df, columns=['Fuel Information.Highway mpg','Engine Information.Engine Statistics.Horsepower'])
df
```

### Linear regression

- Perform linear regression where x = horsepower and y = mpg
- What is the coefficient (slope) of your model? What does this number mean?
- According to your model, what is `y` when `x=1`.
- Show the regression line on a scatterplot with the other datapoints.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

mpg = np.array(df['Fuel Information.Highway mpg']).reshape(-1, 1)
hp = np.array(df['Engine Information.Engine Statistics.Horsepower']).reshape(-1, 1)

regr = LinearRegression()
  
regr.fit(hp, mpg)
print(regr.score(hp, mpg))
print(regr.coef_)
```

```python
mpg_pred = regr.predict(np.array([1]).reshape(1,-1))
mpg_pred
```

```python
df.corr()
```

```python
mgp_pred = regr.predict(hp)
plt.scatter(hp, mpg, color ='b')
plt.plot(hp, mgp_pred, color ='k')
  
plt.show()
```
